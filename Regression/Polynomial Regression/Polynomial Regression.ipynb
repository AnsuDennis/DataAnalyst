```python
# Polynomial Linear Regression : Equation y = b0 + b1 * x1 + b2 * x1 ^ 2+ ----- + bn * x1 ^ n where y = dependent variable, x1 = independent variables, b1, b2, --, bn = coefficients, b0 = constant(pt where line touch x axis. min value)
```

```python
#importing libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
#importing the data set
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x= dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
#training linear regression model on the whole dataset
```

```python
from sklearn.linear_model import LinearRegression
linearRegressor = LinearRegression()
linearRegressor.fit(x,y) # train model on data
```

```python
#training polynomial regression model on the whole dataset
```

```python
# step 1 import tool that allows to create matrix of powered x1
```

```python
from sklearn.preprocessing import PolynomialFeatures
polyMatrix = PolynomialFeatures(degree=2) # y = b0 + b1 * x1 + b2 * x1 ^ 2
x_poly = polyMatrix.fit_transform(x)
```

```python
# step 2 build linear regression model with x_poly
```

```python
polyRegressor = LinearRegression()
polyRegressor.fit(x_poly,y) 
```

```python
#visualising the linear regression results
```

```python
plt.scatter(x,y,color ='red')
plt.plot(x,linearRegressor.predict(x),color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color ='red')
plt.plot(x,polyRegressor.predict(x_poly),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# visualising the polynomial regression results for higher resolution
```

```python
polyMatrix2 = PolynomialFeatures(degree=4) # y = b0 + b1 * x1 + b2 * x1 ^ 2
x_poly2 = polyMatrix2.fit_transform(x)
polyRegressor2 = LinearRegression()
polyRegressor2.fit(x_poly2,y) 
plt.scatter(x,y,color ='red')
plt.plot(x,polyRegressor2.predict(x_poly2),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# visualising the polynomial regression results for higher resolution and smoother curve
```

```python
x_grid=np.arange(min(x),max(x),0.1) # instead of taking 1,2,3 we took 1.1,1.2 etc 
x_grid = x_grid.reshape((len(x_grid),1)) 
plt.scatter(x,y,color ='red')
plt.plot(x_grid,polyRegressor2.predict(polyMatrix2.fit_transform(x_grid)),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
